```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_excel("Online retail.xlsx")
df
```

```python
## Handling the data
```

```python
df.duplicated().any()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.isnull().any()
```

```python
## Transform the data
```

```python
transactions = df.iloc[:, 0].dropna().apply(lambda x: x.split(','))
```

```python
transactions = transactions.apply(lambda x: list(set(x)))
```

```python
te = TransactionEncoder()
encoded_array = te.fit(transactions).transform(transactions)
```

```python
df_encoded = pd.DataFrame(encoded_array, columns=te.columns_)
df_encoded
```

```python
## Association Rule
```

```python
freq_items=apriori(df_encoded,min_support=0.01,use_colnames=True)
freq_items
```

```python
rules=association_rules(freq_items,min_threshold=.5,num_itemsets=2)
rules
```

```python
rules[rules["lift"]>1]
```

## Interview Questions:

1.	What is lift and why is it important in Association rules?

Lift measures how much more likely two items are bought together compared to random chance.

Lift > 1 → Items are frequently bought together.

Lift = 1 → No special relationship (random occurrence).

Lift < 1 → Items are rarely bought together.

2. 	What is support and Confidence. How do you calculate them?

Support – Measures how often an item or itemset appears in all transactions.

Support=Transactions containing (X, Y)/Total Transactions

Confidence – Measures how often Y is bought when X is bought.

Confidence=Support(X, Y)/Support(X)

​Why are they important?

Support helps find popular itemsets.

Confidence shows how strongly two items are related

3.	What are some limitations or challenges of Association rules mining?
​
High Computational Cost – Analyzing large datasets requires significant time and memory.

Too Many Rules – The algorithm may generate too many rules, making it hard to find useful insights.

Low Support for Rare Items – Rare but important patterns might not appear due to low support thresholds.

Interpreting Rules – Not all rules are meaningful or actionable for business decisions.
