#### Mean
The mean is the sum of all values divided by the total number of values.
Formula:
Mean=Sum_of_Observations/Number of Observations

#### Median

The median is the middle value in a sorted list.

Formula:

For even number of observations: 

Median=𝑛/2
 
For odd number of observations:

Median=𝑛+1/2

#### Mode

The mode is the value that appears most frequently.

Formula for grouped data:

Mode=l+((f1-f0)/2f1-f0-f2)*h

l = Lower limit of the modal class

h = Class interval size

𝑓1= Frequency of the modal class

𝑓0=Frequency of the preceding class

𝑓2=Frequency of the succeeding 

#### Standard Deviation
Measures how spread out the data is from the mean.

Formula=Sqrt((sqr(observation-mean_data))/no_of_observation)

#### Importance of Mean, Median, and Mode
These values help identify the "center" of a dataset.

Useful for organizing large datasets.

Standard deviation indicates how data points deviate from the mean.

## STANDARDIZATION OF NUMERICAL VARIABLES

Objective: Standardization, or z-score normalization, scales numerical variables to have a mean of 0 and a standard deviation of 1. This process ensures that variables with different scales do not disproportionately influence analytical models. Many machine learning algorithms, like logistic regression and k-nearest neighbors, perform better on standardized data, as it improves convergence and reduces bias.

Concept of Standardization (z-score normalization) Standardization is the process of transforming data so that each feature’s values have:

Mean (mu) of 0 Standard Deviation of 1 The formula for standardization of a value 𝑥 in a numerical column is:

z=x-mean/standard devition

After standardization, each variable will have a uniform scale, which makes it more suitable for analytical models, particularly those sensitive to data scale.

Steps to Standardize Numerical Columns in Python

Calculate Mean and Standard Deviation

Apply the Z-score Formula:

Compare Distributions Before and After Standardization:

Plot histograms for the original and standardized columns side-by-side to show the difference. Observe the mean (should now be close to 0) and standard deviation (should be 1) in the standardized data.

```python
df_standardized=(df1_numeric-df1_numeric.mean())/df1_numeric.std()
df_standardized
```

```python
for col in df1_numeric.columns:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    axes[0].hist(df1_numeric[col], bins=30, color='skyblue', edgecolor='black')
    axes[0].set_title(f'Histogram of {col} (Original Data)')
    axes[1].hist(df_standardized[col], bins=30, color='orange', edgecolor='black')
    axes[1].set_title(f'Histogram of {col} (Standardized Data)')
    plt.show()
```

Interpreting the Before and After Comparisons

Original Data: The histograms for original data will likely show a variety of means and variances,
reflecting each variable's unique scale.

Standardized Data: The standardized data should now have a mean centered at 0 and a consistent spread, with most values falling within -3 and +3 standard deviations.

Uniformity: The variables are now on a uniform scale, making the dataset more suitable for algorithms that assume standardized input.

Handling Outliers: Outliers will be more noticeable after standardization, as they’ll appear as values far from 0 (above +3 or below -3).

Model Suitability: With standardized data, models that rely on distance or assume normal distribution (like SVM, k-means clustering, or PCA) will likely perform better.
